```python
# estimate pi by monte carlo
import random
import matplotlib.pyplot as plt
import numpy as np
# setup
samples = 100000
x_data = [random.random()-0.5 for x in np.arange(0,samples,1)]
y_data = [random.random()-0.5 for x in np.arange(0,samples,1)]

# initialize lists
inside_x = []
inside_y = []
outside_x = []
outside_y = []
for i in np.arange(0,samples,1):
    if x_data[i]**2 + y_data[i]**2 < 0.25:
        inside_x.append(x_data[i])
        inside_y.append(y_data[i])
    else:
        outside_x.append(x_data[i])
        outside_y.append(y_data[i])
        
estimate = 4*(len(inside_x)/(len(inside_x)+len(outside_x)))
plt.title('guess for pi: ' + str(estimate))
plt.plot(inside_x,inside_y,'r.',outside_x,outside_y,'b.')
plt.legend(['inside','outside'])
plt.axis('equal')
plt.show()
```
