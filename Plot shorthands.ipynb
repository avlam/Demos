```python
# modules
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# some set of data to plot

# sin function over the same span
x1 = np.arange(0,2*np.pi,0.1)
y1 = np.sin(x1)

# random data
x2 = [2*np.pi*random.random() for x in np.arange(0,100,1)] # multiplied by 2pi for same domain as x1
y2 = [2*random.random()-1 for x in np.arange(0,100,1)] #multiplied by 2 and subtracted 1 for same domain as y1

# cos data, cause why not
x3 = np.arange(0,2*np.pi,0.1)
y3 = np.cos(x1)
```

```python
# plot random data as a scatter with red dots, sin wave with blue line with star markers, and cos with green dashed line
sin, = plt.plot(x1,y1,color='blue',marker = '*', label='sin')
plt.scatter(x2,y2,marker='.', color='red',label='random')
cos, = plt.plot(x3,y3,color='green', linestyle='--',label='cos')
plt.legend(loc='best')
plt.show()
```

```python
# plot random data as a scatter with red dots, sin wave with blue line with star markers, and cos with green dashed line
plt.plot(x1,y1,'b*-',x2,y2,'r.',x3,y3,'g--')
plt.legend(['sin','random','cos'],loc='best') #note that the list needs to be the same order as input in plot
plt.show()
```

The shorthand for styling lines within the plot command generally goes color followed by markertype followed by linestyle.
Color shorthands work for 'r','g','b','c','m','y','k','w'
Markertype examples:
    star *
    square s
    triangle ^
    upside down triange v
    right triangle >
    left triange <
    circle o
    dot .
    plus +
    cross x
Linestyle examples:
    dotted line :
    dashed line --
    solid line -
    dot-dash alternating .-

For more detail see matlab's documentation on the plot function. Since matplot lib is built to mirror matlab, the shorthand appears to be the same
